```python
import pandas as pd

df = pd.read_csv('tvmarketing.csv')
df.tail()
```

```python
X = df.drop(columns='Sales')
y = df['Sales']
y.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.xlabel('TV marketing')
plt.ylabel('Sales')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X, y)
beta0 = model.intercept_
beta1 = model.coef_
print("beta0:", + beta0)
print("beta1:", + beta1)
```

```python
import numpy as np

xs = np.linspace(0, 300, 100)
ys = beta0 + beta1 * xs
plt.scatter(X, y)
plt.plot(xs, ys, color='red')
plt.xlabel('TV marketing')
plt.ylabel('Sales')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
# !!!!!!!!!!!!!!!!!!!!!!!! use test values instead of the whole dataset in real projects!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
preds = model.predict(X)
mean_absolute_error(y, preds)
```

```python
plt.scatter(X, y-preds)
```

```python
plt.hist(y-preds, bins=20)
plt.show()
```
